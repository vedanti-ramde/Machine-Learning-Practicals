```python
# STEP 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# STEP 2: Load dataset
df = pd.read_csv("/content/drive/MyDrive/ml_dataset/Spotify_Songs_Dataset.csv")

# STEP 3: Check columns (VERY IMPORTANT)
print(df.columns)

# STEP 4: Select relevant features (Spotify features)
X = df[['danceability', 'energy']]

# STEP 5: Handle missing values (if any)
X = X.dropna()

# STEP 6: Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# STEP 7: Apply KMeans
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)

# STEP 8: Add cluster labels
df = df.loc[X.index]   # keep same rows after dropna
df['cluster'] = kmeans.labels_

# STEP 9: Plot clusters
plt.figure()

sns.scatterplot(
    x='danceability',
    y='energy',
    hue='cluster',
    data=df,
    palette='Set1'
)

# STEP 10: Plot centroids
centroids = scaler.inverse_transform(kmeans.cluster_centers_)

plt.scatter(
    centroids[:, 0],
    centroids[:, 1],
    s=300,
    c='black',
    marker='X',
    label='Centroids'
)

# STEP 11: Title & legend
plt.title("Spotify Songs Clustering using K-Means")
plt.legend()

# STEP 12: Show plot
plt.show()
```
